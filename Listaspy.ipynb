Lista

```python
# Lista aleatória
lst1 = ["ana", "maria", "mariana", 9955, 5599, 5995 ,True, "9559", False]
```

```python
# Print da lista
print(lst1)
```

```python
# Acessar iten na lista
print(lst1[2])
```

```python
# Acessar ultimo iten na lista
print(lst1[-1])
```

```python
# Quantiddade de itens na lista
print(len(lst1))
```

```python
# Tipo da lista
print(type(lst1))
```

```python
# Acessar intervalo da lista
print(lst1[2:7])

# Intervalo do primeiro índice até o 7
print(lst1[:7])

# Intervalo do índice 2 até o último
print(lst1[2:])
```

```python
# Verificar se o item está na lista
if 5599 in lst1:
  print("item encontrado")
else:
  print("item não encontrado")
```

```python
# Alterar item da lista
lst1[4] = "beterraba"

print(lst1[4])
print(lst1)
```

```python
# Alterar um intervalo de itens em uma lista
lst1[2:3] = ["lima", "mouse"]

print(lst1[2:4])
print(lst1)
```

```python
# Alterar um intervalo de itens em uma lista com itens a mais
lst1[2:3] = ["lima", "mouse", "panda", "morcego"]

print(lst1[2:4])
print(lst1)
```

```python
tlst1 = list(("maça", "caju", "manga"))
print(tlst1)
```

```python
# Alterar um intervalo de itens em uma lista a menos
lst1[2:4] = ["suco"]

print(lst1[2:4])
print(lst1)
```

```python
# Inserir novo item na lista sem alterar
lst1.insert(5, "feijoada")

print(lst1[5])
print(lst1)
```

```python
# Inserir novo item no final da lista
lst1.append("Computador")

print(lst1[-1])
print(lst1)
```

```python
# Inserir itens de uma lista em outra lista
lst2 = ["garrafa", "copo", "jarra"]
lst3 = ["celular", "TV", "radio"]

print(lst2)
print(lst3)

lst2.extend(lst3)

print(lst2)
```

```python
# Inserir itens de uma tupla em uma lista
tupla1 = ("mesa", "cadeira")

print(tupla1)

lst2.extend(tupla1)

print(lst2)
```

```python
# Remover item edpecifico da lista
lst2.remove("jarra")

print(lst2)
```

```python
# Remover índice edpecifico da lista
lst2.pop(1)

# também funciona com .del(1)

print(lst2)
```

```python
# Remover último item da lista
lst2.pop()

# também funciona com .pop(-1)

print(lst2)
```

```python
# Apaga a lista completa
del lst2
```

```python
# Esvazia a lista
lst3.clear()

print(lst3)
```

```python
# Imprimir itens da lista um por um com loop
for x in lst1:
  print(x)
```

```python
# imprime itens da lista a partir do tamanho da lista
for i in range(len(lst1)):
  print(lst1[i])
```

```python
# Imprimir itens da lista usando while
i = 0
while i < len(lst1):
  print(lst1[i])
  i += 1
```

```python
# Imprime a lista usando compreensão de lista
[print(x) for x in lst1]
```

```python
lstfruta = ["abacate", "banana", "maça", "uva", "goiaba", "acerola", "maracujá", "moto", "copo"]
novalista = []

print(lstfruta)

for x in lstfruta:
  novalista.append(x)

print(novalista)

novalistafruta = lstfruta
```

```python
# Compreensão de lista com condição
novalista = [x for x in lstfruta if "a" in x]

print(novalista)

# novalista = [expressão for item in interavel if Condição == True]
#O valor retornado é uma nova lista, deixando a lista antiga inalterada.
```

```python
# Compreensão de lista com condição
novalista = [x for x in lstfruta if x != "maça"]

print(novalista)
```

```python
novalista = [x for x in range(10)]

print(novalista)

novalista = [x for x in range(10) if x < 5]

print(novalista)
```

```python
# Definir os itens da nova lista em maiúsculas
novalistafrutas = [x.upper() for x in novalistafruta]

print(novalistafrutas)

novalistafrutas = ["olá" for x in novalistafruta]

print(novalistafrutas)
```

```python
# Condição para manipular itens da lista
novalistafrutas = [x if x != "banana" else "goiaba" for x in novalistafruta]

print(novalistafrutas)

# A expressão no exemplo acima diz:
# "Devolva o produto se não for banana; se for banana, devolva a laranja."
```

```python
# Ordenar lista alfanumérica sort()
listaord = ["Laranja", "Morango", "Limão", "Caju", "Maracujá", "Abacaxi"]
print(listaord)

listaord.sort()
print(listaord)
```

```python
# Ordenar lista alfanumérica sort()
listaordn = [100, 50, 25, 35, 10, 582, 12, 32]
print(listaordn)

listaordn.sort()
print(listaordn)
```

```python
# Ordenar lista em ordem decrescente reverse = true
print(listaord)

listaord.sort(reverse=True)
print(listaord)
```

```python
# Ordenar lista em ordem decrescente reverse = true
print(listaordn)

listaordn.sort(reverse=True)
print(listaordn)
```

```python
def minhafuncao(n):
  return abs(n - 50)

listaordn.sort(key=minhafuncao)
print(listaordn)
```

```python
# A classificação que diferencia maiúsculas de minúsculas pode gerar um resultado inesperado
listaord = ["Laranja", "Morango", "limão", "caju", "maracujá", "Abacaxi"]
print(listaord)

listaord.sort()
print(listaord)
```

```python
# Realiza uma ordenação da lista que ignore maiúsculas e minúsculas com str.lower
listaord.sort()
print(listaord)

listaord.sort(key=str.lower)
print(listaord)
```

```python
# Inverta a ordem dos itens da lista independente da ordem alfabética
listaord = ["Laranja", "Morango", "limão", "caju", "maracujá", "Abacaxi"]
print(listaord)

listaord.reverse()
print(listaord)
```

```python
# lista2 = lista1 não gera uma cópia, mas sim um novo ponteiro para a mesma lista
# Use lista1.copy() para criar um objeto independente."
lista1 = [10, 20, 30]
lista2 = [40, 50, 60]

print(lista1)
print(lista2)

lista2 = lista1

print(lista1)
print(lista2)

lista1 = [10, 20, 30, 70, 80, 90]

print(lista1)
print(lista2)
```

```python
# Criar uma cópia da lista copy()
listaord = ["Laranja", "Morango", "limão", "caju", "maracujá", "Abacaxi"]
minhalista = listaord.copy()
print(minhalista)
```

```python
# Criar uma cópia da lista list()
minhalista = list(listaord)
print(minhalista)
```

```python
# Criar uma cópia da lista com o :operador (slice)
minhalista = listaord[:]
print(minhalista)
```

```python
# Unir ou concatenar duas ou mais listas usando o + operador
lista1 = ["a", "e", "i", "o", "u"]
lista2 = [1, 2, 3, 4, 5]

lista3 = lista1 + lista2

print(lista3)
```

```python
# Unir ou concatenar duas ou mais listas
# Adicionando todos os itens da lista2 a lista1 um por um
lista1 = ["a", "e", "i", "o", "u"]
lista2 = [1, 2, 3, 4, 5]

for x in lista2:
  lista1.append(x)

print(lista1)
```

```python
# Adicionar elementos de uma lista a outra lista usando extend()
lista1 = ["a", "e", "i", "o", "u"]
lista2 = [1, 2, 3, 4, 5]

lista1.extend(lista2)
print(lista1)
```

append() -> Adiciona um elemento ao final da lista.

clear()  -> Remove todos os elementos da lista.

copy()   -> Retorna uma cópia da lista.

count()  -> Retorna o número de elementos com o valor especificado.

extend() -> Adiciona os elementos de uma lista (ou qualquer iterável) ao final da lista atual.

index()  -> Retorna o índice do primeiro elemento com o valor especificado.

insert() -> Adiciona um elemento em uma posição específica.

pop()    -> Remove o elemento em uma posição específica (ou o último, por padrão).

remove()  -> Remove o item com o valor especificado.

reverse() -> Inverte a ordem da lista.

sort()    -> Ordena a lista.
