Dicionários

```python
dicionario = {
    "marca": "Ford",
    "modelo": "ka",
    "ano": "2020"
}
```

Dicionários são usados ​​para armazenar valores de dados em pares chave:valor

Um dicionário é uma coleção ordenada, modificável e que não permite duplicados

Os dicionários são escritos com chaves e possuem chaves e valores

```python
print(dicionario)
```

```python
print(dicionario["marca"])
```

```python
print(len(dicionario))
```

```python
print(type(dicionario))
```

Você pode acessar os itens de um dicionário fazendo referência ao nome da sua chave, dentro de colchetes

```python
# Obtenha o valor da chave "ano"
dicionario = {
    "marca": "Ford",
    "modelo": "ka",
    "ano": "2020"
}

anocarro = dicionario["ano"]

print(anocarro)
```

Existe também um método chamado get() que lhe dará o mesmo resultado

```python
anocarro1 = dicionario.get("ano")

print(anocarro1)
```

O keys() retornará uma lista com todas as chaves do dicionário

```python
chave = dicionario.keys()

print(chave)
```

Adicione um novo item ao dicionário original e veja que a lista de chaves também é atualizada

```python
dicionario ["cor"] = "azul"

print(chave)
```

O values() retornará uma lista com todos os valores do dicionário

```python
valor = dicionario.values()

print(valor)
```

A lista de valores é uma visualização do dicionário, o que significa que quaisquer alterações feitas no dicionário serão refletidas na lista de valores

```python
dicionario["motor"] = "1.0"

print(valor)
```

O items() retornará cada item de um dicionário como tuplas em uma lista

```python
lista = dicionario.items()

print(lista)
```

A lista retornada é uma visualização dos itens do dicionário, o que significa que quaisquer alterações feitas no dicionário serão refletidas na lista de itens

```python
dicionario["cambio"] = "manual"

print(lista)
```

Para determinar se uma chave específica está presente em um dicionário, use in

```python
if "modelo" in dicionario:
  print("Modelo está em dicionario")
```

Você pode alterar o valor de um item específico consultando o nome da sua chave

```python
dicionario["ano"] = "2022"

print(dicionario)
```

O update() atualizará o dicionário com os itens do argumento fornecido

O argumento deve ser um dicionário ou um objeto iterável com pares chave:valor

```python
dicionario.update({"ano": 2025})

print(dicionario)
```

Existem vários métodos para remover itens de um dicionário

O pop() remove o item com o nome de chave especificado

```python
print(dicionario)

dicionario.pop("modelo")

print(dicionario)
```

O popitem() remove o último item inserido (em versões anteriores à 3.7, um item aleatório é removido)

```python
print(dicionario)

dicionario.popitem()

print(dicionario)
```

A del remove o item com o nome de chave especificado

```python
print(dicionario)

del dicionario["motor"]

print(dicionario)
```

A del também pode excluir o dicionário completamente

```python
print(dicionario)

del dicionario

print(dicionario)

# Vai dar erro por ter deletado dicionario
```

O clear() esvazia o dicionário

```python
dicionario = {
    "marca": "Ford",
    "modelo": "ka",
    "ano": "2020"
}

print(dicionario)

dicionario.clear()

print(dicionario)
```

Você pode percorrer um dicionário usando um for

Ao percorrer um dicionário em um loop, o valor retornado são as chaves do dicionário, mas também existem métodos para retornar os valores

```python
dicionario = {
    "marca": "Ford",
    "modelo": "ka",
    "ano": "2020",
    "cor": "azul",
    "cambio": "manual",
    "motor": "1.0"
}
```

```python
# Imprime todos os nomes das chaves no dicionário, um por um
for x in dicionario:
  print(x)
```

```python
# Imprime todos os valores do dicionário, um por um
for x in dicionario:
  print(dicionario[x])
```

Você também pode usar o values() para retornar valores de um dicionário

```python
for x in dicionario.values():
  print(x)
```

Você pode usar o keys() para retornar as chaves de um dicionário

```python
for x in dicionario.keys():
  print(x)
```

Percorre as chaves e os valores usando o items()

```python
for x, y in dicionario.items():
  print(x, y)
```

Não é possível copiar um dicionário simplesmente digitando dicionario2 = dicionario1, porque: dicionario2 será apenas uma referência a dicionario1, e as alterações feitas em dicionario 1 serão automaticamente feitas também em dicionario2

Existem maneiras de fazer uma cópia, uma delas é usar o método Dictionary integrado copy()

```python
print(dicionario)

dicionario2 = dicionario.copy()

print(dicionario2)
```

Outra forma de fazer uma cópia é usar a função integrada dict()

```python
print(dicionario2)

dicionario1 = dict(dicionario2)

print(dicionario1)
```

Um dicionário pode conter outros dicionários, isso é chamado de dicionário aninhado

```python
familia = {
    "filho1" : {
        "nome" : "messsi",
        "idade" : "38"
    },

    "filho2" : {
        "nome" : "ronaldo",
        "idade" : "40"
    },

    "filho3" : {
        "nome" : "neymar",
        "idade" : "34"
    }
}

print(familia)
```

Ou, se você quiser adicionar três dicionários a um novo dicionário

```python
filho1 = {
    "nome" : "messsi",
    "idade" : "38"
}

filho2 = {
    "nome" : "ronaldo",
    "idade" : "40"
}

filho3 = {
    "nome" : "neymar",
    "idade" : "34"
}

familia1 = {
    "filho1" : filho1,
    "filho2" : filho2,
    "filho3" : filho3
}

print(familia1)
```

Para acessar itens de um dicionário aninhado, você usa o nome dos dicionários, começando pelo dicionário externo

```python
print(familia["filho2"]["nome"])
```

Você pode percorrer um dicionário usando items()

```python
for x, a in familia.items():
  print(x)

  for y in a:
    print(y + ':', a[y])
```
