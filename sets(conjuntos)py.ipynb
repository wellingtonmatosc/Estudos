set ou conjuntos

Um conjunto(set) é uma coleção que não é ordenada, imutável e não indexada

Os itens definidos não podem ser alterados, mas você pode remover itens e adicionar novos itens

Os conjuntos são escritos com chaves

```python
conjunto1 = {"maça", "banana", "laranja"}
print(conjunto1)
```

Os conjuntos não estão ordenados, portanto, não é possível garantir a ordem em que os itens aparecerão

Os itens do conjunto não são ordenados, não podem ser alterados e não permitem valores duplicados

```python
# Não são permitidas duplicatas
# Valores duplicados serão ignorados
conjunto2 = {"maça", "banana", "laranja", "banana"}
print(conjunto1)

# Os valores True e 1são considerados o mesmo valor em conjuntos e são tratados como duplicados
conjunto2 = {"maça", "banana", "laranja", "banana", True, 1, 2}
print(conjunto2)

# Os valores False e 0são considerados o mesmo valor em conjuntos e são tratados como duplicados
conjunto3 = {"maça", "banana", "laranja", "banana", False, 1, 0}
print(conjunto3)
```

Os itens do conjunto podem ser de qualquer tipo de dados

```python
conjunto1 = {"maçã", "banana", "limão"}
conjunto2 = {1, 5, 7, 9, 3}
conjunto3 = {True, False, False}

print(conjunto1)
print(conjunto2)
print(conjunto3)
```

```python
# Um conjunto pode conter diferentes tipos de dados
conjunto1 = {"maçã", "banana", 1, 5, 7, 9, False, True}
print(conjunto1)
```

```python
# Tipo de dados do conjunto
print(type(conjunto1))
```

Não é possível acessar itens em um conjunto referenciando um índice ou uma chave

Mas você pode percorrer os itens do conjunto usando um for loop, ou verificar se um valor específico está presente em um conjunto, usando a in

```python
# Percorra o conjunto e imprime os valores
for x in conjunto1:
  print(x)
```

```python
# Verificar item no conjunto
print("maçã" in conjunto1)
print("maca" in conjunto1)

print("banana" not in conjunto1)
print("banan" not in conjunto1)
```

Uma vez criado um conjunto, você não pode alterar seus itens, mas pode adicionar novos itens.

Para adicionar um item a um conjunto, utilize o add()

```python
conjunto1.add("melancia")
print(conjunto1)
```

Para adicionar itens de outro conjunto ao conjunto atual, use o update()

```python
conjunto1 = {"pera", "melancia", "goiaba"}
conjunto2 = {"carro", "moto", "avião"}

print(conjunto1)
print(conjunto2)

conjunto1.update(conjunto2)

print(conjunto1)
```

O objeto no update() não precisa ser um conjunto, pode ser qualquer objeto iterável (tuplas, listas, dicionários etc.)

Para remover um item de um conjunto, use o método remove() ou o discard()

```python
conjunto1.remove("goiaba")
print(conjunto1)

#Se o item a ser removido não existir, remove()ocorrerá um erro
```

Você também pode usar esse pop() para remover um item, mas ele removerá um item aleatório, então você não pode ter certeza de qual item será removido.

```python
print(conjunto1)

cinjunto2 = conjunto1.pop()

print(conjunto2)
print(conjunto1)
```

O clear() esvazia o conjunto

```python
conjunto3 = {"pera", "melancia", "goiaba", "carro", "moto", "avião"}
print(conjunto3)

conjunto3.clear()
print(conjunto3)
```

A del apagará o conjunto completamente

```python
conjunto3 = {"pera", "melancia", "goiaba", "carro", "moto", "avião"}
print(conjunto3)

del conjunto3
# print(conjunto3) # Remover esta linha, pois 'conjunto3' foi deletado e causaria um NameError.
```

Você pode percorrer os itens definidos usando um for

```python
for x in conjunto1:
  print(x)
```

Existem diversas maneiras de unir dois ou mais conjuntos em Python

```python
  conjunto1 = {"banana", "maçã", "laranja", "banana", "limão"}
  conjunto2 = {"melancia", "laranja", "limão", "uva", "mamão"}

  conjunto4 = {1, 2, 3, 2, 1}
  conjunto5 = {4, 5, 6, 2, 1}
```

O método union() e update() combina todos os itens de ambos os conjuntos

```python
# Método union()
print(conjunto1)
print(conjunto2)

conjunto3 = conjunto1.union(conjunto2)
print(conjunto3)
```

```python
# Você pode usar o | em vez do union() e obterá o mesmo resultado
print(conjunto1)
print(conjunto2)

conjunto3 = conjunto1 | conjunto2
print(conjunto3)
```

```python
# Unir vários conjuntos com o union()
print(conjunto1)
print(conjunto2)
print(conjunto4)
print(conjunto5)

conjunto3 = conjunto1.union(conjunto2, conjunto4, conjunto5)
print(conjunto3)
```

```python
# Utilizado |para unir dois conjuntos
print(conjunto1)
print(conjunto2)
print(conjunto4)
print(conjunto5)

conjunto3 = conjunto1 | conjunto2 | conjunto4 | conjunto5
print(conjunto3)
```

O update() insere todos os itens de um conjunto em outro

```python
conjunto4.update(conjunto5)
print(conjunto4)

# Ambos union() e update() excluirão quaisquer itens duplicados
```

O intersection() mantém APENAS os duplicados

O intersection() retornará um novo conjunto, que conterá apenas os itens presentes em ambos os conjuntos

```python
print(conjunto1)
print(conjunto2)

conjunto3 = conjunto1.intersection(conjunto2)
print(conjunto3)
```

Você pode usar o & em vez do método e obterá o mesmo resultado

```python
print(conjunto1)
print(conjunto2)

conjunto3 = conjunto1 & conjunto2
print(conjunto3)
```

O intersection_update() também manterá APENAS os duplicados, mas alterará o conjunto original em vez de retornar um novo conjunto

```python
conjunto4 = {10, 20, 30, 40, 50}
conjunto5 = {40, 50, 60, 70, 80}

print(conjunto4)
print(conjunto5)

conjunto4.intersection_update(conjunto5)

print(conjunto4)
```

O difference() mantém os itens do primeiro conjunto que não estão presentes nos outros conjuntos

O método retornará um novo conjunto que conterá apenas os itens do primeiro conjunto que não estão presentes no outro conjunto

```python
conjunto4 = {10, 20, 30, 40, 50}
conjunto5 = {40, 50, 60, 70, 80}

print(conjunto4)
print(conjunto5)

conjunto3 = conjunto4.difference(conjunto5)

print(conjunto3)
```

Você pode usar o - em vez do método e obterá o mesmo resultado

```python
conjunto4 = {10, 20, 30, 40, 50}
conjunto5 = {40, 50, 60, 70, 80}

print(conjunto4)
print(conjunto5)

conjunto3 = conjunto4 - conjunto5

print(conjunto3)
```

O difference_update() manterá os itens do primeiro conjunto que não estão no outro conjunto, mas modificará o conjunto original em vez de retornar um novo conjunto

```python
conjunto4 = {10, 20, 30, 40, 50}
conjunto5 = {40, 50, 60, 70, 80}

print(conjunto4)
print(conjunto5)

conjunto4.difference_update(conjunto5)

print(conjunto4)
```

O symmetric_difference() mantém todos os itens, EXCETO os duplicados

O symmetric_difference() manterá apenas os elementos que NÃO estão presentes em ambos os conjuntos

```python
conjunto4 = {10, 20, 30, 40, 50}
conjunto5 = {40, 50, 60, 70, 80}

print(conjunto4)
print(conjunto5)

conjunto3 = conjunto4.symmetric_difference(conjunto5)

print(conjunto3)
```

Você pode usar o ^ em vez do método e obterá o mesmo resultado

```python
conjunto4 = {10, 20, 30, 40, 50}
conjunto5 = {40, 50, 60, 70, 80}

print(conjunto4)
print(conjunto5)

conjunto3 = conjunto4 ^ conjunto5

print(conjunto3)
```

O symmetric_difference_update() também manterá todos os itens, exceto os duplicados, mas modificará o conjunto original em vez de retornar um novo conjunto

```python
conjunto4 = {10, 20, 30, 40, 50}
conjunto5 = {40, 50, 60, 70, 80}

print(conjunto4)
print(conjunto5)

conjunto4.symmetric_difference_update(conjunto5)

print(conjunto4)
```

frozenset É uma versão imutável de um conjunto

Assim como os conjuntos, ele contém elementos únicos, não ordenados e imutáveis

Diferentemente dos conjuntos (sets), elementos não podem ser adicionados ou removidos de um frozenset

```python
x = frozenset({"maçã","melancia","uva"})
print(x)
print(type(x))
```

Ser imutável significa que você não pode adicionar ou remover elementos. No entanto, conjuntos congelados suportam todas as operações não mutáveis ​​de conjuntos

Método ----- Operador ----- Descrição

copy() 	Retorna uma cópia superficial (shallow copy) do frozenset.

difference()	-	Retorna um novo frozenset com a diferença entre os conjuntos.

intersection()	&	Retorna um novo frozenset com a interseção (elementos em comum).

isdisjoint()		Retorna True se os dois conjuntos não possuírem nenhum elemento em comum.

issubset()	<= / <	Retorna True se o frozenset for um subconjunto (ou subconjunto próprio) de outro.

issuperset()	>= / >	Retorna True se o frozenset for um superconjunto (ou superconjunto próprio) de outro.

symmetric_difference()	^	Retorna um novo frozenset com a diferença simétrica (elementos que não estão em ambos).

union()	|	Retorna um novo frozenset contendo a união de todos os elementos.

Python possui um conjunto de métodos integrados que você pode usar em conjuntos

add() Adiciona um elemento ao conjunto.

clear() Remove todos os elementos do conjunto.

copy() Retorna uma cópia do conjunto.

difference() - Retorna um novo conjunto com a diferença entre dois ou mais conjuntos.

difference_update() -= Remove do conjunto original os itens que também estão presentes em outro conjunto especificado.

discard() Remove o item especificado (não gera erro se o item não existir).

intersection() & Retorna um novo conjunto que é a interseção de dois ou mais conjuntos.

intersection_update() &= Remove os itens deste conjunto que não estão presentes no(s) outro(s) conjunto(s).

isdisjoint() Retorna se dois conjuntos têm uma interseção ou não (True/False).

issubset() <= Retorna True se todos os itens deste conjunto estiverem presentes no outro conjunto.
< Retorna True se for um subconjunto próprio (contido em um conjunto maior).

issuperset() >= Retorna True se todos os itens de outro conjunto estiverem presentes neste conjunto.
> Retorna True se for um superconjunto próprio (contém um conjunto menor).

pop() Remove e retorna um elemento aleatório do conjunto.

remove() Remove o elemento especificado (gera erro se o item não existir).

symmetric_difference() ^ Retorna um conjunto com as diferenças simétricas de dois conjuntos.

symmetric_difference_update() ^= Atualiza o conjunto original com as diferenças simétricas entre ele e outro.

union() | Retorna um novo conjunto contendo a união de vários conjuntos.

update() |=  Atualiza o conjunto atual com a união entre ele e outros.

Métodos que terminam em _update modificam o conjunto original diretamente em vez de retornar um novo.

discard() vs remove(): Use discard() se você não quer que seu código pare caso o item que você está tentando deletar não exista no conjunto.
